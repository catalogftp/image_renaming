```python
#@title 7. Price & Weight

import numpy as np

price_df = pd.DataFrame()

price_df['sku'] = df_get_data['sku']
price_df['carat_weight'] = df_get_data['weight']
price_df['weight_carat'] = price_df['carat_weight']
price_df['shape'] = df_catalog_others['shape']
price_df['gemstone'] = df_catalog_others['gemstone']
price_df['vendor'] = df_catalog_others['vendor']

# Define a function to calculate tax based on conditions
def calculate_tax(row):
    if row['shape'] in ['Carving', 'Carved'] and 'Coral' not in row['gemstone']:
        return 0.03
    elif 'Coral' in row['gemstone']:
        return 0.05
    elif row['gemstone'] == 'Pearl':
        return 0.03
    else:
        return 0.0025

# Add the 'tax' column using the custom function
price_df['tax'] = price_df.apply(calculate_tax, axis=1)

def custom_round(price):
    if price <= 1000:
        return round(price / 10) * 10
    elif price <= 5000:
        return round(price / 50) * 50
    elif price <= 100000:
        return round(price / 100) * 100
    elif price <= 400000:
        return round(price / 500) * 500
    elif price <= 699999:
        return round(price / 1000) * 1000
    else:
        return 700000

price_df['price'] = ((df_get_data['price_per_carat'] * price_df['tax'])+df_get_data['price_per_carat']) * price_df['carat_weight']

# Apply custom rounding function to 'price' column
price_df['price'] = price_df['price'].apply(custom_round)

# Assuming price_df is a pandas DataFrame containing the price and gemstone columns
price_df['price_per_carat'] = np.where((price_df['price'] == 700000) & (price_df['gemstone'] == 'Alexandrite'), 500000,
                                       np.where(price_df['price'] == 700000, 0, price_df['price'] / price_df['carat_weight']))

# Round off the values in the 'price_per_carat' column to the nearest integer
price_df['price_per_carat'] = price_df['price_per_carat'].round(0)

def callfor(row):
  if row['price'] > 699999:
    return 'Call for Price'
  elif row['price'] == 0:
    return 'Coming Soon'
  else:
    return ''

price_df['callfor'] = price_df.apply(callfor, axis=1)

# Assuming price_df is a pandas DataFrame containing the price column
#price_df['callfor'] = np.where(price_df['price'] > 699999, 'Call for Price', '')



# Assuming price_df is a pandas DataFrame containing the price column
price_df['price_range'] = np.where(price_df['price'] == 700000, np.nan, price_df['price'])

# Convert 'Call for Price' to NaN
price_df['price_range'] = pd.to_numeric(price_df['price_range'], errors='coerce')

# Apply custom rounding function to 'price_range' column
price_df['price_range'] = price_df['price_range'].apply(lambda x: 500 * round(x/500) if not np.isnan(x) else x)

# Ensure price_range is not greater than price
price_df['price_range'] = np.where(price_df['price_range'] > price_df['price'], price_df['price_range'] - 500, price_df['price_range'])

# Convert NaN back to 'Call for Price'
price_df['price_range'] = np.where(price_df['price_range'].isna(), 'Call for Price', price_df['price_range'])


# Assuming price_df is a pandas DataFrame containing the price column
price_df['price_per_carat_range'] = np.where(price_df['price'] == 700000, np.nan, price_df['price_per_carat'])

# Convert 'Call for Price' to NaN
price_df['price_per_carat_range'] = pd.to_numeric(price_df['price_per_carat_range'], errors='coerce')

# Apply custom rounding function to 'price_range' column
price_df['price_per_carat_range'] = price_df['price_per_carat_range'].apply(lambda x: 500 * round(x/500) if not np.isnan(x) else x)

# Ensure price_range is not greater than price
price_df['price_per_carat_range'] = np.where(price_df['price_per_carat_range'] > price_df['price_per_carat'], price_df['price_per_carat_range'] - 500, price_df['price_per_carat_range'])

# Convert NaN back to 'Call for Price'
price_df['price_per_carat_range'] = np.where(price_df['price_per_carat_range'].isna(), 'Call for Price', price_df['price_per_carat_range'])

# Assuming price_df is a pandas DataFrame containing the price and vendor columns
price_df['offers'] = np.where(price_df['price'] == 700000, 'Call for Price', '')

# Adding condition for vendor GP
price_df.loc[(price_df['vendor'] == 'GP') & (price_df['price'] < 700000), 'offers'] = 'Express Dispatch'

price_df['product_category_type'] = np.where(price_df['price'] > 40000, 'P','MP')

# Check if the value is an integer or a float
def custom_round(value):
    if isinstance(value, int):
        return round(value / 0.90)
    elif isinstance(value, float):
        return round(value / 0.90, 2)
    else:
        return value  # Return the original value if it's neither an int nor a float

price_df['weight_ratti'] = price_df['carat_weight'].apply(custom_round)
price_df['weight_ratti_type_price'] = price_df['weight_ratti']
price_df['weight_sort1'] = price_df['carat_weight']*100
price_df['weight_grams'] = round(price_df['carat_weight']/5,2)
price_df['round_of_ratti'] = np.floor(price_df['weight_ratti']/0.25)*0.25

def call_text(row):
    if row['callfor'] == 'Call for Price':
        return 'This is an exclusive and rare category of the gemstone. The price for this category starts from USD 30,000. Please share your contact details and our ' + row['gemstone'] + '  expert will call you to share the exact price.'
    else:
        return ''

price_df['call_for_price_text'] =  price_df.apply(call_text, axis=1)


price_df['approx_weight_range'] = np.floor(price_df['weight_carat']/0.25)*0.25

price_df.drop(columns=['tax'], inplace=True)
price_df.drop(columns=['gemstone'], inplace=True)
price_df.drop(columns=['shape'], inplace=True)
price_df.drop(columns=['vendor'], inplace=True)
```

```python
#@title 14. Check Images Via HTML



# Open the Google Sheet and get the first worksheet
sheet_id = '1Q6ED8SwCXDMluK0DgDPVPuAeq8coLIhHqkPl7Gn6zUs'
sheet = gc.open_by_key(sheet_id)
worksheet = sheet.get_worksheet(0)

# Convert the sheet to a DataFrame
data = worksheet.get_all_records()
df_data = pd.DataFrame(data)

from google.colab import files
from IPython.display import HTML


df = pd.DataFrame(df_data)

# Function to generate HTML for each SKU
def generate_sku_html(sku_group):
    sku_html = "<h1>{}</h1>".format(sku_group['sku'].iloc[0])  # Assuming SKU is the same for each group

    # Collect unique shape, cutting style, and gemstone
    shape = sku_group['shape'].iloc[0]
    cutting_style = sku_group['cutting_style'].iloc[0]
    gemstone = sku_group['gemstone'].iloc[0]

    # Group images and videos
    images = []
    videos = []
    for index, row in sku_group.iterrows():
        if row['image'].endswith(('.jpg', '.jpeg', '.png')):
            images.append(row['image'])
        elif row['image'].endswith('.mp4'):
            videos.append(row['image'])

    # Generate HTML for images
    image_html = ''.join(['<img src="{}" alt="{}" width="250">'.format(image, sku_group['sku'].iloc[0]) for image in images])

    # Generate HTML for video thumbnails
    video_html = ''
    if videos:
        video_html = '<h3>Videos:</h3>'
        for video in videos:
            video_html += """
            <img src="{}" alt="{}" width="250" style="margin-right: 10px;">
            """.format(video + '?thumbnail', sku_group['sku'].iloc[0])

    # Combine images and video thumbnails horizontally
    sku_html += """
    <p>Shape: {}</p>
    <p>Cutting_Style: {}</p>
    <p>Gemstone: {}</p>
    <div style="display:flex; flex-wrap: wrap;">
        {}
    </div>
    {}
    """.format(shape, cutting_style, gemstone, image_html, video_html)

    sku_html += "<hr>"

    return sku_html

# Group DataFrame by SKU and generate HTML content for each SKU
sku_groups = df.groupby('sku')
html_content = """
<!DOCTYPE html>
<html>
<head>
  <title>Product Details</title>
</head>
<body>
"""

for sku, group in sku_groups:
    html_content += generate_sku_html(group)

html_content += """
</body>
</html>
"""

# Save HTML content to a file
with open('product_details.html', 'w') as f:
    f.write(html_content)

# Download the HTML file
files.download('product_details.html')
```
